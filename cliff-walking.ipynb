<details>
  <summary markdown='span'>
  ğŸ’¡ Tuple unpacking?
  </summary>

  Bir fonksiyon tuple dÃ¶ndÃ¼rÃ¼rse, tuple'Ä±n farklÄ± elemanlarÄ±nÄ± hemen farklÄ± deÄŸiÅŸkenlere kaydedebilirsiniz.

  ```python

    """UzunluÄŸu `a` ve geniÅŸliÄŸi `b` olan dikdÃ¶rtgenin
  ```
